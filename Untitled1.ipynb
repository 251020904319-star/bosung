```python
import math

r = float(input("Nhập bán kính r: "))
if r <= 0:
    print("Bán kính phải lớn hơn 0!")
else:
    S = math.pi * r**2
    print("Diện tích hình tròn là:", round(S, 2))
```

```python
year = int(input("Nhập một năm bất kỳ: "))
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    is_leap = True
else:
    is_leap = False
if is_leap:
    print(f"{year} là năm nhuận.")
else:
    print(f"{year} không phải là năm nhuận.")
```

```python

count_negative = 0
count_positive = 0
count_zero = 0
numbers = []
print("Vui lòng nhập 10 số nguyên:")
for i in range(10):
    num = int(input(f"Nhập số thứ {i+1}: "))
    numbers.append(num)
for num in numbers:
    if num < 0:
        count_negative += 1
    elif num > 0:
        count_positive += 1
    else:
        count_zero += 1
print(f"Số lượng số âm: {count_negative}")
print(f"Số lượng số dương: {count_positive}")
print(f"Số lượng số bằng 0: {count_zero}")
```

```python
n = int(input("Nhập một số nguyên dương n: "))
is_prime = True
if n <= 1:
    is_prime = False
else:
    # Kiểm tra các ước số từ 2 đến căn bậc hai của n
    for i in range(2, int(n**0.5) + 1):
        if (n % i) == 0:
            is_prime = False
            break
if is_prime:
    print(f"{n} là số nguyên tố.")
else:
    print(f"{n} không phải là số nguyên tố.")
```
